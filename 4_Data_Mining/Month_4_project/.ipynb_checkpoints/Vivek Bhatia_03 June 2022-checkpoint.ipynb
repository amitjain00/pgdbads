```python
def remove_outlier(col):
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
for i in bank:
    LL,UL=remove_outlier(bank[i])
    bank[i]=np.where(bank[i]>UL,UL,bank[i])
    bank[i]=np.where(bank[i]<LL,LL,bank[i])
```

```python
bank.boxplot(figsize=(10,10))
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
k_means=KMeans(n_clusters=2,random_state=1)
```

```python
k_means.fit(bank_scaled)
```

```python
k_means.labels_
```

```python
k_means.inertia_
```

```python
k_means=KMeans(n_clusters=3,random_state=1)
k_means.fit(bank_scaled)
k_means.inertia_
```

```python
wss=[]
for i in range(1,15):
    KM=KMeans(n_clusters=i,random_state=1)
    KM.fit(bank_scaled)
    wss.append(KM.inertia_)
```

```python
wss
```

```python
plt.plot(range(1,15),wss);
plt.grid()
plt.xlabel('clusters')
plt.ylabel('wss')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
from sklearn import metrics
```

```python
scores = []
k_range = range(2, 15)

for k in k_range:
    km = KMeans(n_clusters=k, random_state=1)
    km.fit(bank_scaled)
    scores.append(metrics.silhouette_score(bank_scaled, km.labels_))
    
scores
```

```python
plt.plot(range(2,15),scores);
plt.grid()
plt.xlabel('clusters')
plt.ylabel('silhouette_score')
plt.show()
```
