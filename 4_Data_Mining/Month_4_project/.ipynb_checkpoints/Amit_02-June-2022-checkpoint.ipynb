### Outlier treatment:

#### First of all define the function:

```python
def update_outlier(col):
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
for i in bank_df:
    LL,UL=update_outlier(bank_df[i])
    bank_df[i]=np.where(bank_df[i]>UL,UL,bank_df[i])
    bank_df[i]=np.where(bank_df[i]<LL,LL,bank_df[i])
```

```python
## Box Plot for all Numeric Fields and check if Outliers have been removed
fig_dims = (7, 6)
fig = plt.subplots(figsize=fig_dims)
sns.boxplot(data=bank_df)
plt.xticks(rotation=60)
```
