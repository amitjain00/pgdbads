### Problem 1A - 2: Perform a one-way ANOVA on Salary with respect to Education. State whether the null hypothesis is accepted or rejected based on the ANOVA results.

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import _get_covariance, anova_lm
```

```python
df_salaryData.columns
```

```python
formula = 'Salary ~ (Education)'
model = ols(formula, df_salaryData).fit()
aov_table = anova_lm(model)
aov_table
```

```python
print("Since the P value is less than alpha value 0.05, so we REJECT the NULL Hypothesis, which says, \
that mean Salary is same across all Education Qualification level, which means there are differences \
in mean salary based on Education  ")
```

### Problem 1A - 3: Perform a one-way ANOVA on Salary with respect to Occupation. State whether the null hypothesis is accepted or rejected based on the ANOVA results.

```python
formula = 'Salary ~ (Occupation)'
model = ols(formula, df_salaryData).fit()
aov_table = anova_lm(model)
aov_table
```

```python
print("Since the P value is Greater than alpha value 0.05, so we FAIL TO REJECT the NULL Hypothesis, which says, \
that Mean Salary is same accross all Occupation level  \
Which means , Mean salary is same for all Occupation categories")
```

### Problem 1B - 2: Perform a two-way ANOVA based on Salary with respect to both Education and Occupation (along with their interaction Education*Occupation). State the null and alternative hypotheses and state your results. How will you interpret this result?

As we confirmed from the interaction plot that there is interactions between education and Occupation levels, So we would conduct a two way anova to check the interaction between Education and occupation.

```python
formula = 'Salary ~ (Education) + (Occupation) + (Education):(Occupation)'
model = ols(formula, df_salaryData).fit()
aov_table = anova_lm(model)
(aov_table)
```

```python
#pca=PCA(random_state=123)
pca = PCA(n_components=17, random_state=123)
df_college_PCA=pca.fit_transform(df_college_scaled)
df_college_PCA
```

### 2.6 Perform PCA and export the data of the Principal Component (eigenvectors) into a data frame with the original features

```python
#Extract eigen vectors
pca.components_
```

```python
#Check the eigen values
#Note: This is always returned in descending order
pca.explained_variance_
```

```python
#Check the explained variance for each PC
#Note: Explained variance = (eigen value of each PC)/(sum of eigen values of all PCs)
pca.explained_variance_ratio_
```

```python
#Create a dataframe containing the loadings or coefficients of all PCs
df_extracted_loadings = pd.DataFrame(pca.components_.T, 
                                     columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13',
                                         'PC14','PC15','PC16','PC17'],
                                    index = df_college_scaled.columns)
df_extracted_loadings
```

```python
#Create a scree plot
plt.figure(figsize=(10,5))
sns.lineplot(y=pca.explained_variance_ratio_ ,x=range(1,18),marker='o')
plt.xlabel('Number of Components',fontsize=10)
plt.ylabel('Variance Explained',fontsize=10)
plt.title('Scree Plot',fontsize=12)
plt.grid()
plt.show()
```

```python
#Check as to how the original features matter to each PC
#Note: Here we are only considering the absolute values
plt.figure(figsize = (20,30))
for i in range(len(df_extracted_loadings.columns)):
    plt.subplot(9,2,i+1)
    abs(df_extracted_loadings[df_extracted_loadings.columns[i]]).T.sort_values(ascending = False).plot.bar()
    plt.yticks(np.arange(0,1.2,.2))
    plt.xticks(rotation=30)
    plt.title('Abs. loadings of {}'.format(df_extracted_loadings.columns[i]))
    plt.tight_layout()
```

```python
#Compare how the original features influence various PCs
plt.figure(figsize = (15,15))
sns.heatmap(abs(df_extracted_loadings), annot = True, cmap = 'Blues',fmt = '.2f');
```

```python
#Choose the PCs basis cumulative explained variance
df_final_PCA = df_extracted_loadings[['PC1','PC2', 'PC3', 'PC4', 'PC5','PC6', 'PC7', 'PC8']]
df_final_PCA
```
